```python
import requests
from bs4 import BeautifulSoup
from lxml import etree

# URL de la página web
url='https://es.wikipedia.org/wiki/Turismo_en_España'

# Realizar la solicitud HTTP y obtener el contenido HTML
response = requests.get(url)
html = response.text

# Crear el objeto BeautifulSoup con analizador 'html.parser'
soup = BeautifulSoup(html, 'lxml')

for table in soup.find_all('table'):
    print(table.get('class'))

tables=soup.find_all('table')
#table=[]
table=soup.find_all('table',class_='wikitable')

print(table[1])


```

```python
filas = table[1].find_all('tr') #es str tengo que poner text

filas[1].find_all('td')[0].text
```

```python
data = []    # lista de listas, matriz


for f in filas:
    
    tmp = []
    
    elem = f.find_all('td')
    
    for e in elem:
        
        tmp.append(e.text.strip().split('[')[0])   #bucle de bucle
        
    data.append(tmp)
        

data
```

```python
import pandas as pd

df = pd.DataFrame(data[1:-1]) #desde segunda celda a la penúltima, df 

#df
```

```python
import pandas as pd

data = [
    [1998, '41,89'],
    [2000, '50,33'],
    [2001, '49,50'],
    [2002, '50,30'],
    [2003, '50,90'],
    [2004, '52,40'],
    [2005, '55,91'],
    [2007, '58,67'],
    [2008, '57,19'],
    [2009, '52,18'],
    [2010, '52,68'],
    [2011, '56,70'],
    [2012, '57,90'],
    [2013, '60,60'],
    [2014, '64,99'],
    [2015, '68,14'],
    [2016, '75,56'],
    [2017, '81,79'],
    [2018, '82,77'],
    [2019, '83,70'],
    [2020, '19,00'],
    [2021, '31,10']
]

df = pd.DataFrame(data, columns=['años', 'Mill_turistas'])

# Convertir los valores en la columna "Mill_turistas" a tipo string
df["Mill_turistas"] = df["Mill_turistas"].astype(str)

# Reemplazar las comas por puntos en la columna "Mill_turistas"
df["Mill_turistas"] = df["Mill_turistas"].str.replace(",", ".").astype(float)

df
```
